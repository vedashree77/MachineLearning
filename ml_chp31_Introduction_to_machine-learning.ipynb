```python
import sklearn
from sklearn.linear_model import LinearRegression
model = LinearRegression()
x = [[1],[2],[3],[4]]
y = [2,4,6,8]
model.fit(x, y)
prediction = model.predict([[6]])
print("Prediction for input 6 is : ", prediction)
```

```python
import sklearn
from sklearn.linear_model import LinearRegression
model = LinearRegression()
x = [[1],[2],[3],[4]]
y = [2,4,6,8]
model.fit(x, y)
user_value = float(input("Enter a number to predict : "))
prediction = model.predict([[user_value]])
print("Prediction is ", prediction[0])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y, color = 'purple', label = 'Data Points')
plt.plot(x,model.predict(x), color= 'pink', label = "Regression Line")
plt.legend(loc = 'center left', bbox_to_anchor = (1,0.5))
plt.title("Linear Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.tight_layout()
plt.show()
```

```python
import sklearn
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
model = LinearRegression()
weights = np.array([1,2,3,4]).reshape(-1,1)
prices = np.array([30,60,90,120])
model.fit(weights, prices)
user_value = float(input("Enter a number to predict : "))
prediction = model.predict([[user_value]])
print("Prediction is ", prediction[0])
plt.scatter(weights,prices, color = 'purple', label = 'Data Points')
plt.plot(weights,model.predict(weights), color= 'pink', label = "Regression Line")
plt.legend(loc = 'center left', bbox_to_anchor = (1,0.5))
plt.title("Linear Regression Line")
plt.xlabel("Weights")
plt.ylabel("Prices")
plt.tight_layout()
plt.show()
```
